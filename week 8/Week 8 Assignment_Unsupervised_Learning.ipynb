## K-means Clustering
Pada pembahasan kali ini akan diuji 2 metode pemilihan nilai cluster (K) yang terbaik, mendekati distribusi pada label anotator.
1. Metode Elbow
2. Via-Score plot

### Metode Elbow

```python
from sklearn.cluster import KMeans
inertia = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(df_scalling.values)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(20, 10))

sns.lineplot(x=range(1, 11), y=inertia, color='#000087', linewidth = 4)
sns.scatterplot(x=range(1, 11), y=inertia, s=300, color='#800000',  linestyle='--')
plt.title('Elbow Method - Finding K value Clusters')
plt.xlabel('The Amount of K')
plt.ylabel('Intertia / WSS')
```

```python
#Dari hasil diatas elbow nya dipilih pada angka 4,
# karena titik awal sebelum titik selanjutnya memiliki gap yang konstan atau
# titik setelah lekukan tajam sebelum mendatar
from sklearn.cluster import KMeans
kmeans_elbow = KMeans(n_clusters=3, random_state=0)
kmeans_elbow.fit(df_scalling.values)
```

```python
#taruh hasil k-means elbow method ke df dengan nama kolom cluster_elbow
df['cluster_elbow'] = kmeans_elbow.labels_
```

### 2. Via Score Plot

```python
# Yellowbrick tidak dapat digunakan di Python 3.12 dalam lingkungan ini karena modul distutils tidak tersedia.
# Kita akan menggunakan plot manual KMeans untuk menentukan nilai k terbaik.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import time

# 1. Inisialisasi list untuk menyimpan inertia dan fit time
inertia = []
fit_times = []
k_range = range(1, 11)

# 2. Proses training dan pencatatan waktu
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    
    start_time = time.time()
    kmeans.fit(df_scalling.values)
    end_time = time.time()
    
    inertia.append(kmeans.inertia_)
    fit_times.append(end_time - start_time)

# 3. Setup canvas utama (untuk plot Inertia / Distortion)
fig, ax1 = plt.subplots(figsize=(10, 6))

# Set grid background abu-abu tipis khas Yellowbrick
ax1.grid(True, which='both', linestyle='-', linewidth=0.5, color='lightgray')
ax1.set_facecolor('#f7f7f7')

# Plot garis biru utama (Distortion/Inertia)
sns.lineplot(x=list(k_range), y=inertia, marker='D', color='#1f77b4', linewidth=2, ax=ax1, label='distortion')
ax1.set_xlabel('k', fontsize=11)
ax1.set_ylabel('distortion', fontsize=11)
ax1.set_xticks(k_range)

# 4. Membuat sumbu Y kedua (untuk Fit Time) di sebelah kanan
ax2 = ax1.twinx()
# Plot garis hijau putus-putus (Fit Time)
sns.lineplot(x=list(k_range), y=fit_times, marker='o', color='#98df8a', linestyle='--', linewidth=1.5, ax=ax2, label='fit time')
ax2.set_ylabel('fit time (seconds)', fontsize=11)
ax2.grid(False) # Matikan grid kedua agar tidak tabrakan

# 5. Menambahkan garis putus-putus vertikal hitam pada k=3 (Nilai optimal)
k_optimal = 3
ax1.axvline(x=k_optimal, color='black', linestyle='--', linewidth=1.5, label=f'k = {k_optimal}')

# 6. Menggabungkan Legend dari kedua sumbu
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

# 7. Judul dan Tampilan
plt.title('Distortion Via-Score Plot for K-Means Clustering', fontsize=12, pad=15)
plt.tight_layout()
plt.show()
```

```python
# Dari hasil diatas K-nya dipilih pada angka 3,
# karena titik distorsinya bertemu pada nilai 3 di sumbu x
from sklearn.cluster import KMeans
kmeans_via = KMeans(n_clusters=3, random_state=0)
kmeans_via.fit(df_scalling.values)
```

### Catatan
Yellowbrick tidak dapat digunakan di Python 3.12 pada lingkungan ini karena modul `distutils` tidak tersedia. Sebagai gantinya, kita menggunakan plot manual dari `KMeans` untuk menentukan nilai `k` terbaik.
